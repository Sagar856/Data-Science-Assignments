```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cutlets = pd.read_csv('/content/Cutlets.csv')
```

```python
cutlets.head()
```

```python
# We are going to conduct a 2-tailed t-test on 2 independent samples
# We need to check weather the mean of both samples are different and is there any significance difference between two samples

# Step-1:

# Make two Hypothesis one contradicting to other,
# Null hypothesis is what we want to prove
# Null Hypothesis :- Ho : mue1 = mue2
# Alternate Hypothesis :- Ha : mue1 != mue2

# Step-2:

# Decide a cut-off value
# significance 5%
# alpha = 0.05
# As it is two-tailed test
# alpha/2 = 0.025
```

```python
# Applying descriptive statistics
cutlets.describe()
```

```python
cutlets.info()
```

```python
# Checking null values
cutlets.isnull().sum()
```

```python
# Checking for duplicate values
cutlets[cutlets.duplicated()].shape
```

```python
# Plotting the data

sns.boxplot(data = cutlets)
```

```python
plt.subplots(figsize = (9,6))
plt.subplot(121)
plt.hist(cutlets['Unit A'], bins = 15)
plt.title("Unit A")
plt.subplot(122)
plt.hist(cutlets['Unit B'], bins = 15)
plt.title('Unit B')
```

```python
plt.figure(figsize = (6,4))
labels = ['Unit A', 'Unit B']
sns.distplot(cutlets['Unit A'], kde =True)
sns.distplot(cutlets['Unit B'], hist = True)
plt.legend(labels)
```

```python
# Plotting Q-Q plot to check weather the distribution follows normal distribution or not.

sm.qqplot(cutlets['Unit A'], line = 'q')
plt.title('Unit A')
sm.qqplot(cutlets['Unit B'], line = 'q')
plt.title('Unit B')
```

```python
# Step-4
# Compare Evidances with Hypothesis using t-statistics

p_value = stats.ttest_ind(cutlets['Unit A'], cutlets['Unit B'], alternative = 'two-sided')
p_value
```

```python
# Compare p_value with alpha (significance level)
# if p_value is not equal to alpha then... we fail to reject null hypothesis
# if p_value is equal to alpha then....we reject null hypothesis

alpha = 0.025
p_value = 0.477

# p_value > alpha

# Hence we fail to reject null hypothesis, i.e there is no significance difference in shape of two samples
```

```python
buyratio = pd.read_csv('/content/BuyerRatio.csv')
```

```python
buyratio.head()
```

```python

# We are going to conduct a Test of independancy i.e Chi-squared test with contingency table.
# We need to check weather the proportion of any of these samples are same or different

# Step-1
# Make two hypothesis one contradicting to other
# Null hypothesis is what we want to prove
# Null hypo :- Ho = There is no association or dependency between the gender based buyer ratios across the regions
# Alternate hypo :- Ha = There is significant association or dependency between gender based buyer ratios.

# Step-2
# Decide a cut-off value
# Significance = 5%
# alpha = 0.05

# Step-3
# Collect Evidences
```

```python
table = [[50,142,131,70],
         [435,1523,1356,750]]
```

```python
# Applying Chi-squared contingency table to convert observed values into expected values

stat, p, dof, exp = stats.chi2_contingency(table)
print(stat,"\n", p,"\n", dof,"\n", exp)
```

```python
observed = np.array([50,142,131,70,435,1523,1356,750])
expected = np.array([42.76531299, 146.81287862,131.11756787,72.30424052,442.23468701,1518.18712138,1355.88243213,747.69575948])
```

```python
# Step-4
# Compare Evidences with hypothesis

statistics, p_value = stats.chisquare(observed, expected, ddof = 3)
print('p_value:', p_value)
```

```python
# Compare p_value with alpha (Significance level)
# If p_value is not equal to alpha....then we failed to reject null hypothesis
# If p_value is equal to alpha ....then we reject null hypothesis
```

```python
# Interpreting p_values

p_value = 0.809
alpha = 0.05

# p_value != alpha

# Hence we fail to reject null hypothesis
# i.e There is no significant association between male and female genders across the regions.
```

```python
# Checking value counts in data

print(cusorder['Phillippines'].value_counts(),'\n',cusorder['Indonesia'].value_counts(),'\n',cusorder['Malta'].value_counts(),'\n',cusorder['India'].value_counts())
```

```python
# Creating contingency table

contingency_table = [[271,267,269,280],
                     [29,33,31,20]]
```

```python
# Applying chi-squared contingency values to convert observed values into expected values

stat, p, dof, exp = stats.chi2_contingency(contingency_table)
print(stat,'\n',p,'\n',dof,'\n',exp)
```

```python
# Defining expected values and Observed values

observed = np.array([271,267,269,280,29,33,31,20])
expected = np.array([271.75, 271.75, 271.75, 271.75, 28.25,  28.25,  28.25,  28.25])
```

```python
# Compare evidances with Hypothesis using t-statistic

t_stat, p_value = stats.chisquare(observed, expected, ddof = dof)
print('t-stat:',t_stat,'\n', 'p_value:',p_value)
```

```python
# Compare p_value with alpha
# If p_value is not equal to alpha...then we fail to reject null hypothesis
# If p_value is equla to alpha ...then we reject null hypothesis
```

```python
# Interpreting p_value

p_value = 0.4254
alpha = 0.05

# p_value != alpha

# Hence we fail to reject null hypothesis
# i.e there are no defect varies with the country.
```
